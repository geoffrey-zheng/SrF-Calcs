```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("XBMolLongVelProfileAvg5Shots.txt", skiprows=8, names=['index', 'count', 'rate'], delimiter=', ')
counts_slowing = df['count'].to_numpy()
counts_slowing = np.mean(counts_slowing.reshape([-1, 5]), axis=1)

df = pd.read_csv("NoXBMolLongVelProfileAvg5Shots.txt", skiprows=8, names=['index', 'count', 'rate'], delimiter=', ')
counts_no_slowing = df['count'].to_numpy()
counts_no_slowing = np.mean(counts_no_slowing.reshape([-1, 5]), axis=1)


plt.plot(counts_slowing)
plt.plot(counts_no_slowing)
```

```python
df = pd.read_csv("AlteredXBMolLongVelProfileAvg5Shots.txt", skiprows=8, names=['index', 'count', 'rate'], delimiter=', ')
counts = df['count'].to_numpy()
counts = np.mean(counts.reshape([-1, 5]), axis=1)
counts_no_slowing = counts[::2]
counts_slowing = counts[1::2]

plt.plot(counts_no_slowing)
plt.plot(counts_slowing)
```

```python
df = pd.read_csv("AlteredXBMolLongVelProfile-06282024.txt", skiprows=8, names=['index', 'count', 'rate'], delimiter=', ')
counts = df['count'].to_numpy()
# counts = np.mean(counts.reshape([-1, 5]), axis=1)
counts_no_slowing = counts[::2]
counts_slowing = counts[1::2]

plt.plot(counts_no_slowing, label='no slowing')
plt.plot(counts_slowing, label='slowing')
plt.legend()
plt.show()
```

```python
df = pd.read_csv("LongVelProfileAltXB472MHzRed2MHzStep-07092024.txt", skiprows=8, names=['index', 'count', 'rate'], delimiter=', ')
counts = df['count'].to_numpy()
# counts = np.mean(counts.reshape([-1, 5]), axis=1)
counts_no_slowing = counts[::2]
counts_slowing = counts[1::2]
vel = (572-np.linspace(100, 500, len(counts_no_slowing)))*1e6*787e-9
plt.plot(vel, counts_no_slowing, label='no slowing')
plt.plot(vel, counts_slowing, label='slowing')
plt.xlabel("Molecule velocity (m/s)")
plt.ylabel("Fluorescence Counts (arb.)")
plt.legend()
plt.show()
```
